```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(black['Amount (USD)'], bins=50, kde=True, color="skyblue")
plt.title('Distribution of Transaction Amounts (USD)')
plt.xlabel('Amount (USD)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Money Laundering Risk Score', data=black, palette="viridis")
plt.title('Money Laundering Risk Score Distribution')
plt.xlabel('Risk Score')
plt.ylabel('Count')
plt.show()
```

```python
# Group by 'Country' to analyze the average transaction amount per country
country_amount = black.groupby("Country")["Amount (USD)"].mean().sort_values(ascending=False)
```

```python
country_amount
```

```python
# Group by 'Industry' to analyze the average transaction amount per industry
industry_amount = black.groupby("Industry")["Amount (USD)"].mean().sort_values(ascending=False)
```

```python
industry_amount
```

```python
# Group by 'Money Laundering Risk Score' to analyze the average transaction amount per risk score
risk_amount = black.groupby("Money Laundering Risk Score")["Amount (USD)"].mean()
```

```python
risk_amount
```

```python
# Filter high-risk and high-value transactions (risk score 8-10 and amount in top 25%)
high_risk_transactions = black[(black["Money Laundering Risk Score"] >= 8) &
                              (black["Amount (USD)"] >= black["Amount (USD)"].quantile(0.75))]
```

```python
high_risk_transactions
```

```python
# Plot average Amount (USD) by Country
plt.figure(figsize=(12, 6))
sns.barplot(x=country_amount.index, y=country_amount.values, palette="viridis")
plt.title('Average Transaction Amount by Country')
plt.xlabel('Country')
plt.ylabel('Average Amount (USD)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot average Amount (USD) by Industry
plt.figure(figsize=(12, 6))
sns.barplot(x=industry_amount.index, y=industry_amount.values, palette="magma")
plt.title('Average Transaction Amount by Industry')
plt.xlabel('Industry')
plt.ylabel('Average Amount (USD)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot average Amount (USD) by Risk Score
plt.figure(figsize=(10, 6))
sns.lineplot(x=risk_amount.index, y=risk_amount.values, marker='o', color="crimson")
plt.title('Average Transaction Amount by Money Laundering Risk Score')
plt.xlabel('Risk Score')
plt.ylabel('Average Amount (USD)')
plt.show()
```

```python
# Display the high-risk, high-value transactions
high_risk_transactions_summary = high_risk_transactions[['Country', 'Industry', 'Amount (USD)', 'Money Laundering Risk Score']].groupby(['Country', 'Industry']).size().reset_index(name='Count')
high_risk_transactions_summary.sort_values(by='Count', ascending=False).head(10)
```
